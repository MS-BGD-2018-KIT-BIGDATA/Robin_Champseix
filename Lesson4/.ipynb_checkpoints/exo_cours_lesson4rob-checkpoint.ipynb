```python
import requests
from bs4 import BeautifulSoup
import re
import pandas as pd
from itertools import chain
from multiprocessing import Pool, TimeoutError
import numpy as np
import urllib.request
import json
```

```python
jsonIbu = "https://www.open-medicaments.fr/api/v1/medicaments?query=ibuprofene"
```

```python
with urllib.request.urlopen(jsonIbu) as url:
        data = json.loads(url.read())
```

```python
codeList = list(map(lambda x : data[x]["codeCIS"], range(len(data))))
```

```python
urlMedicament = "https://www.open-medicaments.fr/api/v1/medicaments/"
liste_Url_Medicament = list(map(lambda x : urlMedicament + x, codeList))
```

```python
def get_Infos(urlMedicament):
    jsonData = requests.get(urlMedicament).json()
    
    code = jsonData["codeCIS"]
    
    regex_Nom = r'.* [0-9]{1}'
    chaine_Nom = jsonData["denomination"]
    nom = re.search(regex_Nom, chaine_Nom)[0]
    nom = nom[:-1]
    
    dosage = jsonData["compositions"][0]["substancesActives"][0]["dosageSubstance"]
    
    regex_Comprimé = r"[0-9]* *(comprimé)"
    chaine_Comprimé = jsonData["presentations"][0]['libelle']
    if re.search(regex_Comprimé, chaine_Comprimé):
        nb_Comprimé = re.search(regex_Comprimé, chaine_Comprimé)[0]
    else : 
        nb_Comprimé = "Gel"
   
    
    regex_Age = r"[0-9]* (ans)"
    chaine =jsonData["indicationsTherapeutiques"]
    if re.search(regex_Age, chaine):
        contreIndicationAge = re.search(regex_Age, chaine)[0]
    else : 
        contreIndicationAge = "Aucune"
        
    prix = jsonData["presentations"][0]['prix']
    
    return code, nom, dosage, nb_Comprimé,contreIndicationAge, prix
```

```python
df = pd.DataFrame(list(map(lambda x : get_Infos(x), liste_Url_Medicament)),
                  columns=["code","nom", "dosage", "nb_Comprimé","contreIndicationAge", "prix"])
```
